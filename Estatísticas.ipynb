Amostragem Sistemática:

```python
from math import ceil
```

```python
#Criação das variáveis para representar a população a amostra e o valor de k
populacao = 150
amostra = 15
k = ceil(populacao/amostra)
k
```

```python
#Definição do valor randômico para inicializar a amostra, iniciando em 1 até k+1
r = np.random.randint(low= 1 , high= k+1 , size = 1)
r
```

```python
#Criamos um for para somar os próximos valores, baseado no primeiro valor r criado acima 
acumulador = r[0]
sorteados = []
for i in range(amostra):
  #print(acumulador)
  sorteados.append(acumulador)
  acumulador += k
print(sorteados)
```

```python
base_amostragem = base.loc[sorteados]
base_amostragem
```
