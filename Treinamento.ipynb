```python
Pessoas = pd.read_excel('Pessoas.xlsx')
Pessoas.head()
```

```python
filtro = Pessoas.groupby(['Idade']).Salario.mean()
filtro.plot(label='Senso', color= '#444444', linestyle='--', marker='.', linewidth=1) #markers + . 
plt.style.use('ggplot') #plt.xkcd()
#plt.xkcd()
plt.ylabel('Salario (BRL)')
plt.xlabel('Idades')
plt.title('Media de Salario x Idade')
plt.grid(True)
plt.tight_layout()
plt.savefig('plot.png')
plt.show()



```

**Visualisando Dados com Matplotlib**

Dataframes permitem um grande meio para explorar e analizar dados tabulares, mas as vezes uma imagem vale mais do que  mil  linhas e colunas. A biblioteca Matplotlib fornece base para plotar visualizações de dados que podem melhorar muito sua capacidade de analisar os dados.

Vamos começar com um gráfico de barras simples que mostra a média salarial por idade nos países latino americanos.

```python
latinos_salario = Pessoas[Pessoas.Pais.isin(['Brazil','Argentina','Mexico','Peru','Bolivia','Colombia','Chile','Venezuela','Ecuador','Paraguay'])].groupby(['Idade']).Salario.mean()

```

```python
fig = plt.figure(figsize=(10,3))
latinos_salario.plot.bar(color='purple')
plt.xlabel('Idade')
plt.ylabel('Salário')
plt.title('Media Salarial x Idade (Latinos)')
plt.grid(color='#95a5a6',linestyle='--', linewidth=1,alpha=0.5)
plt.xticks(rotation=90)
plt.show()
```
